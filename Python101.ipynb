### HW คำนวณเวลาเป็นวินาทีของเวลาต่อไปนี้โดยใช้คำสั่ง split() ช่วย (print ออกมาให้สวยงาม)
12:30:15

13:41:07

12:53:15

00:59:25

11:11:11

16:06:09

````python
time_list = [
    "12:30:15",
    "13:41:07",
    "12:53:15",
    "00:59:25",
    "11:11:11",
    "16:06:09"
]

print(f"{'เวลา (HH:MM:SS)':<18} | {'คำนวณเป็นวินาทีได้':<18}")
print("-" * 40)

# วนลูปเพื่อแยกส่วนและคำนวณ
for t in time_list:
    # ใช้ split(':') แยกออกเป็น [ชั่วโมง, นาที, วินาที]
    hours, minutes, seconds = t.split(':')

    # แปลงจาก string เป็น integer แล้วคำนวณหาผลรวมวินาที
    total_seconds = (int(hours) * 3600) + (int(minutes) * 60) + int(seconds)

    # แสดงผลลัพธ์แบบจัดรูปเล่มให้สวยงาม
    print(f"{t:<18} | {total_seconds:>7,} วินาที")
````

## ให้สร้างฟังก์ชั่นคำนวณเวลาเป็นวินาทีของเวลาใดๆ และ print ออกมาให้สวยงาม

````python
def convert_time_to_seconds(time_str):
    # ใช้ split(':') เพื่อแยกชั่วโมง นาที วินาที ออกจากกัน
    hours, minutes, seconds = time_str.split(':')

    # แปลงเป็น integer แล้วคำนวณหาผลรวมวินาที
    total_seconds = (int(hours) * 3600) + (int(minutes) * 60) + int(seconds)

    # แสดงผลลัพธ์แบบจัด Format ให้สวยงามและตรงล็อค
    print(f"เวลา {time_str:<10} คำนวณเป็นวินาทีได้ทั้งหมด: {total_seconds:>7,} วินาที")

    return total_seconds
````

## ให้ สร้าง list ของ เวลา ```['12:30:15','13:41:07',....]``` แล้ววนลูปเรียกฟังก์ชั่นคำนวณเวลาเป็น วินาที

**commit ว่า HW3**

````python
time_list = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]

# ทำเป็นหัวตารางเพื่อความสวยงาม
print("=" * 55)
print(f"{'สรุปการคำนวณเวลาเป็นวินาที':^55}")
print("=" * 55)

# วนลูปเพื่อเรียกใช้งานฟังก์ชันกับทุกค่าใน list
for time_item in time_list:
    convert_time_to_seconds(time_item)

print("=" * 55)
````
