### def คือการกำหนดฟังก์ชัน


```python
def function_name(Input):
    do_something with _Input_ to get _Output_
    return _Output_
```
- def function_name(Input) คือ การประกาศสร้างฟังก์ชัน ตั้งชื่อฟังก์ชัน และกำหนดตัวแปรนำเข้า (Parameter) ไว้รอรับค่า

- do_something... คือ การเขียนคำสั่งหรือกระบวนการทำงานข้างในฟังก์ชัน เพื่อคำนวณหรือจัดการข้อมูลให้ได้ผลลัพธ์

- return _Output_ คือ การส่งค่าผลลัพธ์ที่ได้จากการทำงานในฟังก์ชันออกไป เพื่อให้โค้ดส่วนอื่นนำค่านี้ไปใช้งานต่อ

4.   ส่วน output ```return _Output_```(ขาดได้)

## เราสามารถกำหนดค่า default ให้กับฟังก์ชั่นได้

default(ค่าเริ่มต้น) = ค่าที่ระบบเลือกให้โดยอัตโนมัติ ในกรณีที่ผู้ใช้งานไม่ได้กำหนดหรือระบุค่าอะไรลงไป

input ของ function ใน python มีสองแบบ input ที่จำเป็นต้องใส่ กับ input ที่ไม่จำเป็นต้องใส่ (มีค่า default)

เราต้องเรียง input ที่จำเป็นต้องใส่ขึ้นก่อน

````python
def print_2lines_default(name,surname,ID,grade='F'):
    st = f'ชื่อ {name} นามสกุล {surname} รหัส {ID}'
    print (st)
    st2=f'เกรดวิชา Data Viz >>> {grade}'
    print(st2)
````

````python
print_2lines_default('ธันยาการต์','พวงมาลัย','613020551-8')
````

````python
print_2lines_default('ธันยาการต์','พวงมาลัย','613020551-8','A')
````

# LOOP การวนซ้ำ

for เป็นคำที่ใช้บอก python ว่าเรากำลังเขียน loop
โดย for จะวนดึงสมาชิกจาก ```listA``` มาทำ process ```do_something```

```python
    for each_member in listA :
        do_something
```

#### คำสั่ง for เป็นคำสั่งวนซ้ำที่ใช้ควบคุมการทำงานซ้ำๆ ในจำนวนรอบที่แน่นอน

````python
for i in [1,2,3] :
    o = i**2
    print (f'this member = {i} after process = {o}')
````

### loop in function
#### การวนลูปในฟังก์ชั่น¶

````python
def print_feelings(I,friendS,feeling = 'คิดถึง'):
    for friend in friendS:
        print(f'{I} {feeling} {friend}')
````

````python
list_friends = ['พี่บูม','ดายน์','แตงโม','ฝน','พี่เจ็ท','น้องแคมป์']
````

````python
print_feelings('เจมส์',list_friends)
````

````python
print_feelings('เจมส์',list_friends,feeling='รัก')
````

````python
print_feelings('เจมส์',list_friends,'รัก')
````

### conditional คือ เงื่อนไข /ประโยคเงื่อนไข

##### True / False
##### การวนลูปโดยมีเงื่อนไขต่างๆมาเกี่ยวข้อง โดยมี pattern ดังนี้


```Python
if condition1 :
    do something
elif condition2 :  #elif คือ else if
    do another thing
else:
    do ...
```
    

````python
print(list_friends)
````

````python
for name1 in list_friends :
    for name2 in list_friends :
            print(name1,'รัก',name2)
````

````python
for name1 in list_friends:
    for name2 in list_friends:
        if name1 != name2: #ไม่ปริ้นชื่อคนเดียวกันซ้ำ
            print(name1,'รัก',name2)
````

````python
for name1 in list_friends:
    for name2 in list_friends:
        if name1 == name2: #ปริ้นชื่อคนเดียวกันซ้ำ
            print(name1,'รัก',name2)
````

````python
for name1 in list_friends:
    for name2 in list_friends:
        if name1 == name2:
            'do nothing'
        else:
            print(name1,'รัก',name2)
````
